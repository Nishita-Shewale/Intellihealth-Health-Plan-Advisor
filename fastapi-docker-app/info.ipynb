```python
import csv

# Path to your CSV file
file_path = '/Users/anoushka/Desktop/insurance_plans.csv'
output_path = '/Users/anoushka/Desktop/cleaned_insurance_plans.csv'

# Read the CSV file
with open(file_path, 'r') as file:
    reader = csv.reader(file)
    rows = [row for row in reader]

# Ensure that every row has the same number of columns (109 columns in your case)
# This script will add empty values for extra commas or remove trailing commas
cleaned_rows = []
expected_columns = 109

for row in rows:
    # Remove extra commas (if any), and ensure the row has the expected number of columns
    if len(row) < expected_columns:
        # Pad with empty fields if there are fewer columns than expected
        row.extend([''] * (expected_columns - len(row)))
    elif len(row) > expected_columns:
        # Trim extra columns if there are more columns than expected
        row = row[:expected_columns]
    
    cleaned_rows.append(row)

# Write the cleaned rows to a new CSV file
with open(output_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(cleaned_rows)

print(f"Cleaned CSV saved at {output_path}")
```

```python
row_count = 0
column_count = 0

# Open the cleaned CSV file and count rows and columns
with open(output_path, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        row_count += 1
        # Update column_count with the number of columns in the first row (assuming consistent column count)
        if row_count == 1:
            column_count = len(row)

# Print the row and column counts
print(f"Number of rows: {row_count}")
print(f"Number of columns: {column_count}")
```

```python
columns_to_clean = [
    "SpecialtyDrugMaximumCoinsurance",
    "DEHBInnTier1IndividualMOOP",
    "DEHBInnTier1FamilyPerPersonMOOP",
    "DEHBInnTier1FamilyPerGroupMOOP",
    "HSAOrHRAEmployerContributionAmount"
]

# Define a function to clean monetary columns
def clean_monetary_column(column):
    return (
        column.astype(str)  # Convert all values to string for processing
        .str.replace(r"[^\d.]", "", regex=True)  # Remove non-numeric characters like $, commas, etc.
        .replace("", None)  # Replace empty strings with None (optional)
        .astype(float)  # Convert to float
    )

# Apply the cleaning function to the specified columns
for col in columns_to_clean:
    df_cleaned[col] = clean_monetary_column(df_cleaned[col])

# Preview the cleaned columns
for col in columns_to_clean:
    print(f"Cleaned values in {col}:")
    print(df_cleaned[col].unique())
```
